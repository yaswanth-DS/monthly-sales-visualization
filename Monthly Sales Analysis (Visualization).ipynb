## Bar Chart — Compare months 

```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [18000, 22000, 25000, 24000, 30000, 90000]
plt.bar(months,sales)
plt.title("monthly sales")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

## Line Chart — Trend

```python
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [18000, 22000, 25000, 24000, 30000, 90000]
plt.plot(months, sales, marker='o')
plt.title("Sales Trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

## Histogram — Distribution 

```python
import seaborn as ssn
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [18000, 22000, 25000, 24000, 30000, 90000]
plt.hist(sales, bins=4)
plt.title("Sales distribution")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

## Boxplot — Outliers 

```python
import seaborn as ssn
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [18000, 22000, 25000, 24000, 30000, 90000]
ssn.boxplot(y=sales)
plt.title("Sales outliers")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```
